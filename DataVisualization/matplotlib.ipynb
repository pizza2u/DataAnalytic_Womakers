# IntroduÃ§Ã£o a Matplotlib: Primeiros GrÃ¡ficos, Figuras, Titulos e RÃ³tulos


**Primeiros GrÃ¡ficos**


```python
import matplotlib.pyplot as plt

# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Criando um grÃ¡fico de linha
plt.plot(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Meu Primeiro GrÃ¡fico')
plt.show()
```

**Figuras, TÃ­tulos e RÃ³tulos**

```python
import matplotlib.pyplot as plt

# Criando uma nova figura
# fig representa a figura em que o grÃ¡fico serÃ¡ desenhado
# ax representa o eixo (ou os eixos) dentro da figura onde o grÃ¡fico Ã© plotado
fig, ax = plt.subplots()

# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Criando um grÃ¡fico de linha
ax.plot(x, y)
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title('GrÃ¡fico Personalizado')
plt.show()
```

Ajustar o layout da figura

```python
import matplotlib.pyplot as plt

# Criando uma nova figura
fig, ax = plt.subplots()

# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Criando um grÃ¡fico de linha
ax.plot(x, y)
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title('GrÃ¡fico Personalizado')

# Ajustar o layout
fig.tight_layout()

plt.show()
```

Adicionar uma anotaÃ§Ã£o na figura:



```python
import matplotlib.pyplot as plt

# Criando uma nova figura
fig, ax = plt.subplots()

# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Criando um grÃ¡fico de linha
ax.plot(x, y)
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title('GrÃ¡fico Personalizado')

# Adicionar uma anotaÃ§Ã£o
fig.text(0.5, -0.05, 'Nota: Este Ã© um grÃ¡fico de exemplo', ha='center', fontsize=12)

plt.show()
```

```python
# Criando um grÃ¡fico de linha com anotaÃ§Ã£o
plt.plot(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('GrÃ¡fico com AnotaÃ§Ã£o')

# Adicionando anotaÃ§Ã£o
plt.annotate('Ponto Interessante', xy=(3, 5), xytext=(4, 6),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

**Tipos de GrÃ¡ficos**

AlÃ©m de grÃ¡ficos de linha, o Matplotlib suporta diversos tipos de grÃ¡ficos:

GrÃ¡fico de Barras

```python
# Dados
categorias = ['A', 'B', 'C', 'D']
valores = [4, 7, 1, 8]

# Criando um grÃ¡fico de barras
plt.bar(categorias, valores)
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('GrÃ¡fico de Barras')
plt.show()
```

```python
# Dados
categorias = ['A', 'B', 'C', 'D']
valores = [4, 7, 1, 8]

# Criando um grÃ¡fico de barras
plt.barh(categorias, valores)
plt.ylabel('Categorias')
plt.xlabel('Valores')
plt.title('GrÃ¡fico de Barras')
plt.show()
```

GrÃ¡fico de DispersÃ£o


```python
# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Criando um grÃ¡fico de dispersÃ£o
plt.scatter(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('GrÃ¡fico de DispersÃ£o')
plt.show()
```

# PersonalizaÃ§Ã£o Matplotlib: Subplots, Paletas de Cores e Salvando GrÃ¡ficos

Subplots


```python
# Criando subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Dados
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 9, 16, 25]

# Primeiro subplot
ax1.plot(x, y1, 'b-')  # linha azul
ax1.set_title('GrÃ¡fico 1')
ax1.set_xlabel('Eixo X')
ax1.set_ylabel('Eixo Y')

# Segundo subplot
ax2.plot(x, y2, 'r-')  # linha vermelha
ax2.set_title('GrÃ¡fico 2')
ax2.set_xlabel('Eixo X')
ax2.set_ylabel('Eixo Y')

plt.tight_layout()
plt.show()
```

Paletas de Cores


```python
# Usando uma paleta de cores
plt.style.use('ggplot')

# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Criando um grÃ¡fico de linha
plt.plot(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('GrÃ¡fico com Estilo ggplot')
plt.show()
```

ConfiguraÃ§Ãµes de Eixos


```python
# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Criando um grÃ¡fico de linha com configuraÃ§Ã£o de eixos
plt.plot(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('GrÃ¡fico com ConfiguraÃ§Ãµes de Eixos')

# Configurando os eixos
plt.xlim(0, 6)
plt.ylim(0, 12)
plt.xticks([1, 2, 3, 4, 5], ['A', 'B', 'C', 'D', 'E'])
plt.show()
```

Legendas

```python
# Dados
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 9, 16, 25]

# Criando um grÃ¡fico de linha com legendas
plt.plot(x, y1, label='SÃ©rie 1')
plt.plot(x, y2, label='SÃ©rie 2')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('GrÃ¡fico com Legendas')
plt.legend()
plt.show()
```

Interatividade


```python
from mpl_toolkits.mplot3d import Axes3D

# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
z = [1, 4, 9, 16, 25]

# Criando um grÃ¡fico 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Eixo Z')
ax.set_title('GrÃ¡fico 3D')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
z = [1, 4, 9, 16, 25]

# Criando um grÃ¡fico 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Eixo Z')

# Usar fig.suptitle para o tÃ­tulo geral
fig.suptitle('GrÃ¡fico 3D')

# Ajustar layout para garantir que as legendas dos eixos sejam legÃ­veis
plt.subplots_adjust(left=-0.5, right=0.9, top=0.9, bottom=0.1)

plt.show()
```
