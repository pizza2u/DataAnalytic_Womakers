# Seaborn

O Seaborn é uma biblioteca poderosa de visualização de dados em Python, construída sobre o Matplotlib, que facilita a criação de gráficos estatísticos atraentes e informativos. Com uma interface de alto nível, o Seaborn é ideal para analistas de dados, cientistas de dados e todos aqueles que desejam explorar e comunicar dados de forma visualmente agradável.

```python
# !pip install seaborn -- nÃ£o Ã© necessÃ¡rio no Colab
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

1.  Gráficos com Seaborn

    1.1. Gráficos de Dispersão

Os gráficos de dispersão são úteis para visualizar a relação entre duas variáveis contínuas. Com o Seaborn, podemos criar gráficos de dispersão facilmente usando a função scatterplot().

```python
# Carregar dados de exemplo
data = sns.load_dataset('iris')

# Criar gráfico de dispersão
sns.scatterplot(data=data, x='sepal_length', y='sepal_width', hue='species')
plt.title('Gráfico de Dispersão: Comprimento vs. Largura das Sépalas')
plt.show()
```

1.2. Gráficos de Barras

Os gráficos de barras são ideais para comparar diferentes categorias. Podemos usar a função barplot() para criar gráficos de barras no Seaborn.

```python
# Criar gráfico de barras
sns.barplot(data=data, x='species', y='sepal_length')
plt.title('Comprimento Médio das Sépalas por Espécie')
plt.show()
```

1.3. Histogramas

Os histogramas são, de fato, uma ferramenta poderosa para visualizar a distribuição de variáveis, permitindo que você veja a frequência de diferentes intervalos de valores. A função histplot() do Seaborn facilita muito essa tarefa. Ela combina a funcionalidade de criar histogramas com a opção de adicionar estimativas de densidade, o que pode tornar a visualização mais informativa.

```python
# Criar histograma
sns.histplot(data=data, x='sepal_length', kde=True)
plt.title('Distribuição do Comprimento das Sépalas')  # Corrigido
plt.show()
```

1.4. Boxplots

Os boxplots são uma ferramenta eficaz para visualizar a distribuição de dados, especialmente para identificar outliers e entender a variação dos dados. A função boxplot() do Seaborn facilita a criação desses gráficos.

```python
# Criar boxplot
sns.boxplot(data=data, x='species', y='sepal_length')
plt.title('Boxplot do Comprimento das Sépalas por Espécie')  
plt.show()
```

2. Paletas de Cores

O Seaborn oferece diversas paletas de cores que podem ser usadas para melhorar a estética dos gráficos. Podemos usar a função color_palette() para definir paletas de cores.

2.1. Paletas Padrão

O Seaborn possui paletas padrão como deep, muted, bright, pastel, dark e colorblind.

```python
# Aplicar paleta 'pastel'
sns.set_palette('pastel')
sns.boxplot(data=data, x='species', y='sepal_length')
plt.title('Boxplot com Paleta Pastel')
plt.show()
```

2.2. Paletas Personalizadas

Podemos criar paletas personalizadas usando a função color_palette() com cores definidas manualmente.

```python
# Criar paleta personalizada
custom_palette = sns.color_palette(['#FF6347', '#4682B4', '#3CB371'])
sns.set_palette(custom_palette)
sns.boxplot(data=data, x='species', y='sepal_length')
plt.title('Boxplot com Paleta Personalizada')
plt.show()
```

2.3. Paletas Divergentes

As paletas divergentes são úteis para representar dados que têm um ponto central significativo, como zero.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Criar mapa de calor com paleta divergente
flights = sns.load_dataset('flights')
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')

# Verificar os valores presentes no Ã­ndice 'month'
print(flights_pivot.index)

# Acessar o valor central para o heatmap
central_value = flights_pivot.loc['Jan', 1955]

sns.heatmap(flights_pivot, cmap='coolwarm', center=central_value)
plt.title('Mapa de Calor com Paleta Divergente')
plt.show()
```

3. Temas e Estilos

O Seaborn facilita a aplicação de temas e estilos para tornar os gráficos mais atraentes e consistentes com as preferências estéticas.

3.1. Temas Padrão

Os temas padrão do Seaborn incluem darkgrid, whitegrid, dark, white e ticks.

```python
# Aplicar tema 'whitegrid'
sns.set_theme(style='whitegrid')
sns.boxplot(data=data, x='species', y='sepal_length')
plt.title('Boxplot com Tema Whitegrid')
plt.show()
```

3.2. Ajuste Fino de Estilos

Podemos ajustar elementos individuais de estilo, como contextos e paletas.

```python
# Ajustar contexto para apresentaÃ§Ã£o
sns.set_context('talk')
sns.set_style('darkgrid')
sns.boxplot(data=data, x='species', y='sepal_length')
plt.title('Boxplot com Contexto Talk e Estilo Darkgrid')
plt.show()
```

3.3. Remoção de Eixos

Para uma aparência mais limpa, podemos remover os eixos dos gráficos.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar dataset de exemplo
data = sns.load_dataset('iris')

# Criar grÃ¡fico de boxplot sem eixos
sns.boxplot(data=data, x='species', y='sepal_length')
sns.despine()  # Remove a espinha dorsal do grÃ¡fico

# Remover ticks e labels dos eixos
plt.gca().xaxis.set_ticks([])
plt.gca().yaxis.set_ticks([])
plt.gca().xaxis.set_ticklabels([])
plt.gca().yaxis.set_ticklabels([])

# TÃ­tulo do grÃ¡fico
plt.title('Boxplot sem Eixos')
plt.show()
```
