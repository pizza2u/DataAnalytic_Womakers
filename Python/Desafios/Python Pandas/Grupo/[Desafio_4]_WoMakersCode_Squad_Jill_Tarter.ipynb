#Criando a coluna TotalGoals a partir da soma das colunas HomeGoals e AwayGoals

```python
df['TotalGoals'] = df['HomeGoals'] + df['AwayGoals']
df
```

#Encontrando o valor máximo na coluna TotalGoals

```python
df['TotalGoals'].max()
```

#Localizando o índice em que o valor máximo encontrado está

```python
indice = df[df['TotalGoals'].isin([12])]
indice
```
