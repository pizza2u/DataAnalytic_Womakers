#Importação das bibliotecas pydataset e pandas

```python
from pydataset import data
import pandas as pd
```

#Importação do dataset "women" do pydataset

```python
df = data('women')
df.head()
```

###1. Construa uma nova representação onde os nomes das colunas estão em português, e os valores de altura são convertidos para centímetros e os valores de peso convertidos para quilograma.

Dicas:
1 polegada = 2,54 centímetros
1 libra = 0,45 quilos

```python
df1 = df.rename(columns={'height': 'altura', 'weight': 'peso'})
df1['altura'] = df1['altura'] * 2.54
df1['peso'] = df1['peso'] * 0.45
df1.head()
```

#2. Plote o scatterplot de altura x peso da nova representação.

```python
df2 = df1.plot.scatter(x='altura',
                      y='peso'
                      )
```
