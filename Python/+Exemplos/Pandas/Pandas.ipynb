```python
from datetime import datetime
```

```python
def calcular_idade(data_nascimento):
    data_nascimento = pd.to_datetime(data_nascimento)
    hoje = datetime.now()  # Usa datetime.now() para obter a data atual
    idade = hoje.year - data_nascimento.year - ((hoje.month, hoje.day) < (data_nascimento.month, data_nascimento.day))
    return idade

# Adicionar a coluna 'Idade'
df_concurso['Idade'] = df_concurso['Data de Nascimento'].apply(calcular_idade)

media_idade_por_estado = df_concurso.groupby('Estado')['Idade'].mean()
print("Média de idade por estado:")
for estado, media_idade in media_idade_por_estado.items():
    print(f"{estado}: {media_idade:.0f}")
```
